```python
import requests, bs4
url = 'https://www.bseindia.com/markets/equity/eqreports/topmarketcapitalization.aspx'
headers = {"User-Agent":"Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = bs4.BeautifulSoup(response.text, 'lxml')
```

```python
table=soup.find("table",class_="mGrid")
```

```python
headers=table.find_all("th")
titles=[]
for i in headers:
    title = i.text
    titles.append(title)
print(titles)    
```

```python
pd.DataFrame(columns=titles)
```

```python
data=table.find_all("td")
```

```python
# extracting row data 
row_data=[]
for i in data:
    row = i.text.strip()
    row_data.append(row)
    
```

```python
row_data
j=0
new_list=[]
lst=[]
for i in range(0,len(row_data)):
    if j<3:
        new_list.append(row_data[i].strip())
        j+=1 
    else:
        new_list.append(row_data[i].strip())
        lst.append(new_list)
        new_list=[]
        j=0
```

```python
df = pd.DataFrame(lst, columns=titles)
```

```python
df.head()
```
